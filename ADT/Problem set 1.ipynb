```python
from math import floor

def divide(x: int, y: int):
    if x == 0:
        return 0,0
    z = floor(x / 2)
    q,r = divide(z,y)
    q = 2*q
    r = 2*r
    if x % 2 == 1:
        r = r + 1
    if r >= y:
        r = r - y
        q = q + 1
    return q,r
```

```python
divide(34235235232,3243242)
```

```python
def verifydivide(x,y,a,b):
    if a*y + b == x:
        return True
    else:
        return False
```

```python
import random

def createexample(n: int, u: int):
    '''
    :param n: Number of random examples
    :param u: Upper limit on the number
    :return:
    '''

    for i in range(n):
        x = random.randint(0,u)
        y = random.randint(0,u)
        print(f"""--------------------------------------

x,y are {x},{y}
""")
        a,b = divide(x,y)
        print(f"---> DIVIDE gives q:{a},r:{b}")
        if verifydivide(x,y,a,b):
            print("---> DIVIDE is True")
        else:
            raise Exception(f"Found a False Solution {x},{y},{a},{b}")

    return
```

```python
createexample(100, 10000000)
```
