```python
def fizzbuzz(x: int) -> int:
    x = x -1
    mult_of_3 = x // 3
    mult_of_5 = x // 5
    mult_of_15 = x // 15

    def sum_till(n: int):
        return n * (n + 1) // 2

    total_sum = 3*sum_till(mult_of_3) + 5*sum_till(mult_of_5) - 15*sum_till(mult_of_15)

    return total_sum
```

```python
fizzbuzz(1000)
```

```python
def evenfibsum(x: int) -> int:
    a1 = 1
    a2 = 2
    total_sum = 0
    while a2 < x:
        if a2 % 2 == 0:
            total_sum += a2
        a1, a2 = a2, a1 + a2

    return total_sum
```

```python
evenfibsum(4000000)
```

```python
def collatzlen(x: int) -> int:
    len = 1

    while x != 1:
        if x % 2 == 0: x = x // 2
        else: x = 3*x + 1
        len += 1

    return len

def argmaxcollatz(x: int) -> int:
    maxlen = 1
    maxint = 1

    for i in range(1,x):
        k = collatzlen(i)
        if k > maxlen:
            maxlen = k
            maxint = i

    return maxint
```

```python
argmaxcollatz(1000000)
```
