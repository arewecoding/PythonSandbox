```python
from pydantic import BaseModel, EmailStr, field_validator


class User(BaseModel):
    name: str
    email: EmailStr # Here we are using an in-build validator to validate an email data type.
    account_id: int

    # Here we're defining a function which will become a custom validator for the account_id data type. So every time someone creates an obj -> User, the value for account_id will be passed through this function first.
    # The field_validator decorator tell the pydantic library to use this function as a validator for the 'account_id' data type.
    @field_validator('account_id')
    def validate_account_id(cls, value):
        if value <= 0:
            raise ValueError(f"Account ID must be positive: {value}")
        return value

user = User(name="Jack", email="jack@gmail.com", account_id=1234)
print(user)
```

```python
print(type(user))
print(user.model_dump_json())
```

```python
json_str = '{"name":"Jack","email":"jack@gmail.com","account_id":1234}'
user2 = User.model_validate_json(json_str)
```
